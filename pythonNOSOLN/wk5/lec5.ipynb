# Time complexity analysis using Big-$O$ Notation
Now that we have learned how to write functions, we want to start thinking about how efficient our functions (or algorithms) are. In order to analyze how good your algorithm is in terms of time, we instead count how many constant time operations the function is performing. Then we use a notation called Big-$O$ that represents the worst-case running time of a program.  
We compute the Big-O of an algorithm by counting how many <b>constant time operations</b> an algorithm will take in the worst-case scenario based on the size of the input.

Why use this formalized method?

Timing your CPU is not a very good way to analyze the run time of a program -- your CPU is doing many things. On top of this, who knows what operating system you are using and how this will affect the run time.

What does it mean for a program to be efficient? Among others:

* CPU (time) usage -- time complexity
* memory/disk usage -- space complexity

<b> Time complexity </b> refers to how well a program will <emph> scale</emph>: If I only have a small dataset, my program runs really fast -- but how fast does it run for a dataset that is 10 times larger?  The <b>Big-$O$</b> notation is used to represent the worst case time complexity of an algorithm. The time complexity is analyzed is based on the number of constant time operations performed.

<b> Space complexity </b> describes, similarly, how well a programs memory usage scales with input size.

This lecture is going to be on time complexity only.

Let's quickly revisit this `prime` function

````python
def prime(n):

    is_prime = True
    # loop through the numbers 2 to (n - 1)
    for m in range(2, n):
        # if n is evenly divisible by m, we already know n is not a prime
        # so break out of the loop
        if n % m == 0:
            is_prime = False
            break

    if is_prime:
        return True
    else:
        return False
````

Big-O notation is how we measure how "long it takes" to run a function based on what the input to that function is. It is based on counting the number of <b> constant time operations </b> a function runs.

## Constant time operations
Constant time operations are simply operations who's run time does not change with the input. For example:

* arithmetic operation
* assignments
* boolean tests / branching
* reading/writing a single line

Branching (checking if a condition is T/F, for example) is also constant time.

# Time complexity analysis of a function

Time complexity analysis is done by determining the total number of constant time statements.

Here are a few rules:
1. Constant time statements are simple statements (listed above)

* arithmetic operation
* assignments
* boolean tests
* reading/writing a single line

Their complexity is also called $O(1)$.

2. Loops are multiplied by the number of times the loop is run

```python
for idx in range(10):
    print(idx)
```
    
The print counts as a single constant time operation. It is run ten times. So this piece of code has 10 constant time operations.
    
Similarly, nested loops multiply:

```for ii in range(10):
    for jj in range(20):
        ii * jj
```
        
The inner loop executes an arithmetic operation. That counts as 1. But it's run a total of $10\times20$ times. So in total, this nested for loop is $10\times20\times1 = 200$.

3. If a statement is a function call, the complexity of the statement is the complexity of that function.


4. if-then statements: count whichever block is longer:

```python
if x > 20:
    for i in range(x):
        print(x)
else:
    print("x is small")
```
    

If $x\leq20$, then there is only one print statement ($O(1)$). If not, then a loop is run. So, we only consider the worst case scenario which is that the entire loop runs $x$ times in total.

## Runtime analysis of sum of squares function
Let's look at an example:

````python
def square(lst):
    """ squares all values in a list"""
    for idx in range(len(lst)):
        lst[idx] = lst[idx]**2

def sum_of_squares(lst):
    """Returns the sum of squares of all numbers in a list"""
    square(lst)

    rtn_val = 0
    for val in lst:
        rtn_val = rtn_val + val
    return rtn_val
````

(remember that lists are mutable, so `square(lst)` will update the list object outside of the function).

How long does it take to run sum_of_squares? Or, in other words, how fast is the sum-of-squares algorithm?

````python
a = [0, 1, 2, 3, 4]
````

````python
sum_of_squares(a)
````

Let's count the number of constant time operations.

For size of the input `lst` is `n` (in the above example n = 5)
* First, `square(lst)` is called
    * `square(lst)` will loop through the list and square each entry. --> This is a total of `n` multiplication operations (i.e. `n` constant time operations).
* Then, we set `rtn_val = 0`, which is 1 constant time operation.
* Finally, we loop through the squared version of the list and add to `rtn_val`
    * This is `n` addition operations (another `n` constant time operations).

 In total, there were $n + 1 + n + 1 = 2n + 2$ constant time operations. In other words, the running time of calling the function `sum_of_squares` is proportional to $2n + 2$ where $n$ is the size of the input list.

## Runtime analysis of `prime` funtion
Here is our old prime function that determines if a number `n` is prime or not.
Let's count all the constant operations.

````python
def prime(n):
    is_prime = True
    # loop through the numbers 2 to (n - 1)
    for m in range(2, n):
        # if n is evenly divisible by m, we already know n is not a prime
        # so break out of the loop
        if n % m == 0:
            is_prime = False
            break

    if is_prime:
        print(n, " is a prime! ")
    else:
        print(n, " is not a prime")
````

For an input `n`, in the ``worst case`` --> complete the loop and never break out early:
* 1 constant time operation - setting is_prime to be True
* Loop through the numbers 2 to n and do the following each time
    * check if n %m == 0 --> a constant time operation
    * total is (n - 2) * 1 operations
* check if is_prime is True: 1 constant time operation
* print something: 1 constant time operation

Total:  $1 + (n - 2)*1 + 1 + 1 = n + 2$

## `prime` vs `sum_of_squares`
What is really the point so far? `sum_of_squares` had a run time of $2n + 2$ while `prime` is $n + 1$.

How much does that $2n$ matter compared to the $n$? Or the $+1$ term? Well, not much. Next time you upgrade your computer, the constant time operations might be twice as fast anyways so there's really no difference between something that takes $2n$ and $n$. There's also no real difference between something that has an additional few constant operations (the $+1$ terms) and one that doesn't.

What <emph> does </emph> matter is the fact that using prime on a small number, say $n = 5$ is going to be much quicker than using prime on a number that's 1000-fold larger.

|Function | Input Size | Number of Operations |
| --- | --- | --- |
| `sum_of_squares` | 10 |21 |
| `sum_of_squares` | 100 |201 |
| `sum_of_squares` | 1000 |2002 |
| `sum_of_squares` | 1000000 |200002 |
| `prime` | 10 |11 |
| `prime` | 100 |101 |
| `prime` | 1000 |1001 |
| `prime` | 1000000 |100001 |

Is there really much of a difference between 200002 and 100001 operations? Not in a way that is noticeable to you. What is noticable is that finding out if 1000000 is prime or not will take 10 times longer than finding out if 100000 is prime or not. Finding the sum of squares of a list of length 500 is 100 times longer than a list of size 5.


<b> The main point is </b>

Both `prime` and `sum_of_squares` scale proportional to the size of the input.

Both `prime` and `sum_of_squares` are both <b> $O(n)$ </b> algorithms. Neither one is "better" or considered "faster" than the other. All that we care about is that they both scale proportional to the input size.  


## Runtime analysis of a simple sorting algorithm
Here is another example with a function that sorts a list using the bubble sort algorithm.

Visualize algorithm here
* https://visualgo.net/en/sorting



````python
def bubble_sort(lst):
    """ sorts list in ascending order using bubble sort algorithm"""
    n = len(lst)

    # flag that keeps track if any lst[left] > lst[right]
    swapped = False

    # go through each entry in list
    for left in range(n - 1):

        # check all elements to the right of it
        for right in range(n - left - 1):

            # if it is out of order
            if lst[right] > lst[right + 1]:
                # set swapped flag to True
                swapped = True

                # swap elements
                tmp = lst[right + 1]
                lst[right + 1]  = lst[right]
                lst[right] = tmp

        if not swapped:
            # no swaps were made, then list is sorted
            return
````

````python
a = [898, -9, 12, 1, 2, 4, 0, 3, 431, 23, 21]
bubble_sort(a)
a
````

Here are all the operations of bubble_sort in the <b> worst </b> case
* find the length of the list, constant time = 1
* set swapped to true, constant time = 1
* loop from 0 to n - 1
    * Loop from left to n - 1 -->  another n - 1
        * if statement = 1
        * if we have to swap:
            * set swapped = 1
            * swap elements = 3
* if statement = 1

Total $= 2 + n (n - 1) \times 5 + 1 = 5n^2 - 5n + 3$.

This is the worst case --> it is possible that the list is already sorted! Then all the if statements are skipped and the run time is actually less. But this is what we want to analyze. We want to look at the worst case scenario.

So what's the runtime here? Well we're going to ignore all the constant terms. So our runtime becomes $n^2 - n$.
Which term dominates $n^2$ or $n$ when we have a large enough list to sort? Of course, $n^2$ will.

For lists that are large enough that we'd even want or care to write a sorting algorithm for, the $n$ term will be so small compared to $n^2$ term. So the $n^2$ is what's dominating the runtime here.

This `bubble_sort` algorithm is $O(n^2)$. (That's not a very good algorithm actually).

### Formal definition

So what is Big $O$? It is the asymptotic upper bound on the runtime of a function (algorithm).
<b>Asymptotic</b> here refers to the case of very large inputs.

Here's the super formal math definition, in case it means anything to you: The set of all functions $f$ in $O(g(x))$ if

$f(x) \in O g(x) \iff \exists c, N_0 >0 | \forall N \geq N_0, \frac{|f(x)|}{|g(x)|} \leq c$

This is saying a function $f$ is in $O(g(x))$ if and only if positive numbers $N_0$ and $c$ exist such that $\frac{f}{g}$ is at most $c$. In other words "in the long-run, the function $f$ won't be faster than $g$".

But anyways for our purposes, we just want to start thinking of how efficient an algorithm is in terms of its dominating factor.

# Finding the square root of a number numerically

Let's write a function to find the square root of a number numerically. We will set a threshold on "how close" we want our program to be to the right answer. This is also called an <b> approximation algorithm </b>.

A "close enough" answer is one that is within some constant <emph> epsilon</emph> to the actual answer. We get to decide what epsilon is.

Suppose we are trying to find the square root of $10$. One way to do this would be to look at all numbers between 0 and 10, and check if the square is "close enough" to 10.:

Set epsilon = 0.01

* Is $|(10 - 0^2)| \leq \text{epsilon}$?
* Is $|(10 - 0.001^2)| \leq \text{epsilon}$?
* Is $|(10 - 0.002^2)| \leq \text{epsilon}$?
* Is $|(10 - 0.003^2)| \leq \text{epsilon}$?
* and on and on until you find the x that satisfies $(10 - x^2) \leq \text{epsilon}$ (or your x is larger than 10 which means you missed the square root).

Each time we check a number that is 0.001 larger than the last number we checked. Here, 0.001 is called the <b> step size </b>.

Let's put this into a function.

````python
def square_root(n, epsilon=0.01, step=0.001):
    """ Find the approximate root of n within epsilon accuracy"""

    rtn = 0
    while abs(rtn ** 2 - n) >= epsilon and rtn <= n:
        # check the next number
        rtn = rtn + step

    # when the while loop is exited, either the number we are checking (rtn)
    # is too big (failed) or we found a number rtn that is within
    # epislon of the square root

    if (abs(rtn ** 2 - n) < epsilon):

        print("Square root of %.2f is approximately %.4f  \t(epsilon = %.4f, step = %.4f)"
              % (n, rtn,  epsilon, step))
        return rtn
    else:
        print("Failed to find square root of %.2f \t (epsilon = %.4f, step = %.4f)"
              % (n,  epsilon, step))
    return -1
````

````python
square_root(10, 0.01, 0.0001)
(10 - 3.1607**2)
````

This is an example of an <b>exhaustive</b> or <b> brute-force</b> algorithm. We go through <emph>all</emph> possible answers between 0 and $n$ and find the first one that "works" (satisfies whatever condition we set -- i.e. "close enough").

`prime` defined above is also another brute-force algorithm. It works, but it's not really efficient. Anyways. Let's look at the time complexity of `square_root`.

* rtn = 0, constant time  
* The `while` loop starts with rtn = 0 and in the worst case it stops when rtn > n. Each time we increase `rtn` by the step size. In total, the loop is run $n / 0.01$.  Each time the loop is run
    * `while` loop conditions are checked --> constant time
    * `rtn = rtn + stp` --> constant time (arithmetic operation)
    * In total the loop is run a constant times $n$ and each time it does two constant time operations, so this entire loop is $3n$.
* both clauses of the `if` statement are constant time (print and return) - 2 constant time operations

Total = 1 + 3n + 1.


in Big -$O$ notation, the constants don't matter, $O(3n + 2) = O(n)$.

<b> Conclusion </b>
The `square_root` function is $O(n)$.

Here is a version of square root that also returns a counter for each time the while loop is executed

````python
def square_root2(n, epsilon=0.01, step=0.001):
    """ Find the approximate root of n within epsilon accuracy"""

    rtn = 0
    ctr = 0
    while abs(rtn ** 2 - n) >= epsilon and rtn <= n:
        # check the next number
        rtn = rtn + step
        ctr = ctr + 1

    # when the while loop is exited, either the number we are checking (rtn)
    # is too big (failed) or we found a number rtn that is within
    # epislon of the square root

    if (abs(rtn ** 2 - n) < epsilon):

        print("Square root of %.2f is approximately %.4f  \t(epsilon = %.4f, step = %.4f)"
              % (n, rtn,  epsilon, step))
        return rtn, ctr
    else:
        print("Failed to find square root of %.2f \t (epsilon = %.4f, step = %.4f)"
              % (n,  epsilon, step))
    return -1, ctr
````
